```python
import folium
```

```python
import folium.plugins



m = folium.Map(location=[-29, -49], zoom_start=6)
folium.LatLngPopup().add_to(m)
folium.plugins.Draw().add_to(m);
```

```python
# 2016-12-06 a 2019-12-06

y0, y1 = -33, -24
x0, x1 = -53, -44


points = [
    (y0, x0),
    (y1, x0),
    (y1, x1),
    (y0, x1),
    (y0, x0),
]

folium.PolyLine(points, weight=3, color="red").add_to(m);
```
